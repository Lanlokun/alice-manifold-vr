```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# 1. Setup Model
device = "cuda" if torch.cuda.is_available() else "cpu"
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model_gpt = GPT2LMHeadModel.from_pretrained("gpt2").to(device)

def get_word_surprisal(text):
    inputs = tokenizer(text, return_tensors="pt").to(device)
    with torch.no_grad():
        outputs = model_gpt(**inputs, labels=inputs["input_ids"])
    # Loss here is the negative log-likelihood (surprisal)
    return outputs.loss.item()

# 2. Map Transcription Segments to TRs (374 bins)
tr_surprisal = np.zeros(374)
tr_duration = 2.0  # seconds per fMRI volume

for seg in result['segments']:
    tr_idx = int(seg['start'] / tr_duration)
    if tr_idx < 374:
        # Calculate surprisal for the specific phrase heard in this time window
        s_val = get_word_surprisal(seg['text'])
        tr_surprisal[tr_idx] = s_val

# 3. Interpolate to fill gaps between spoken sentences
from scipy.interpolate import interp1d
nonzero_indices = np.where(tr_surprisal > 0)[0]
f_interp = interp1d(nonzero_indices, tr_surprisal[nonzero_indices], kind='linear', fill_value="extrapolate")
surprisal_final = f_interp(np.arange(374))

print("Surprisal Signal Ready.")
```

```python
import numpy as np
import nibabel as nib
from sklearn.decomposition import PCA
from nilearn import plotting
import matplotlib.pyplot as plt

# --- DYNAMIC PATH CONFIGURATION ---
# Use the BASE_PATH from the previous cell
if 'BASE_PATH' not in locals():
    BASE_PATH = '/home/lanlokun/Documents/Nankai University/FinalYear/THESIS/The Universal Narrative Engine/Datasets/ds002322'

def get_pca_and_mask(sub_id, n_components=50):
    """Loads preprocessed data and returns the PCA object + voxel mask."""
    DERIVATIVES_PATH = os.path.join(BASE_PATH, 'derivatives')
    path = os.path.join(DERIVATIVES_PATH, f'sub-{sub_id}', f'sub-{sub_id}_task-alice_bold_preprocessed.nii.gz')
    
    print(f"Loading PCA data for subject {sub_id} from: {path}")
    img = nib.load(path)
    data = img.get_fdata()
    
    # Reshape to (Time, Voxels)
    flattened = data.reshape(-1, data.shape[3]).T
    
    # Create mask of non-zero variance voxels
    mask_indices = np.where(np.std(flattened, axis=0) > 0)[0]
    masked_data = flattened[:, mask_indices]
    
    # Run PCA
    pca = PCA(n_components=n_components)
    pca.fit(masked_data)
    
    return pca, mask_indices

# 1. Re-generate PCA objects to ensure they exist in memory
print("Re-indexing Subject 18 and 30...")
pca_18, mask_18 = get_pca_and_mask('18')
pca_30, mask_30 = get_pca_and_mask('30')

# 2. Back-project CCA weights through PCA components to Voxel Space
# Note: we use the weights from your most recent successful 'cca_val' fit
voxel_weights_18 = pca_18.components_.T @ cca_val.x_weights_[:, 0]
voxel_weights_30 = pca_30.components_.T @ cca_val.y_weights_[:, 0]

# 3. Map back to 3D Volumes
def build_3d(weights, mask, shape):
    vol = np.zeros(np.prod(shape))
    vol[mask] = weights
    return np.abs(vol).reshape(shape)

spatial_map_18 = build_3d(voxel_weights_18, mask_18, brain_shape)
spatial_map_30 = build_3d(voxel_weights_30, mask_30, brain_shape)

print(f"SUCCESS: Volumes reconstructed. Sub-18 Max: {np.max(spatial_map_18):.4f}")
```
