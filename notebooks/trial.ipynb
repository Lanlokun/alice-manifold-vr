```python
# ==========================================================
# FMRI MANIFOLD PIPELINE
# ==========================================================

import os
import numpy as np
import nibabel as nib
from sklearn.decomposition import TruncatedSVD
from scipy.stats import pearsonr


def load_subject(subject_id, base_path):
    
    func_path = os.path.join(
        base_path,
        "derivatives",
        f"sub-{subject_id}",
        f"sub-{subject_id}_task-alice_bold_preprocessed.nii.gz"
    )
    
    if not os.path.exists(func_path):
        raise FileNotFoundError(f"Cannot find {func_path}")
        
    img = nib.load(func_path)
    data = img.get_fdata()
    
    flattened = data.reshape(-1, data.shape[3]).T
    
    return flattened
```

```python
def preprocess_fmri(data, max_timepoints=500):

    if data.shape[0] > max_timepoints:
        data = data[:max_timepoints]

    voxel_std = np.std(data, axis=0)
    mask = voxel_std > 1e-10

    reduced = data[:, mask]

    centered = reduced - np.mean(reduced, axis=0)

    return centered, mask
```

```python
def compute_manifold(data, n_components=3):

    pca = TruncatedSVD(n_components=n_components)

    manifold = pca.fit_transform(data)

    return manifold, pca
```

```python
def project_subject(data, mask, pca):

    filtered = data[:, mask]

    centered = filtered - np.mean(filtered, axis=0)

    manifold = pca.transform(centered)

    return manifold
```

```python
def trajectory_similarity(traj1, traj2):

    min_len = min(len(traj1), len(traj2))

    t1 = traj1[:min_len]
    t2 = traj2[:min_len]

    corr_x, _ = pearsonr(t1[:,0], t2[:,0])
    corr_y, _ = pearsonr(t1[:,1], t2[:,1])
    corr_z, _ = pearsonr(t1[:,2], t2[:,2])

    return np.mean([corr_x, corr_y, corr_z])
```

```python
BASE_PATH = "../data/ds002322-download"

print("Loading Subject 18")

sub18 = load_subject(18, BASE_PATH)
sub18_processed, mask = preprocess_fmri(sub18)

manifold18, pca = compute_manifold(sub18_processed)

print("Loading Subject 22")

sub22 = load_subject(22, BASE_PATH)

manifold22 = project_subject(sub22, mask, pca)

similarity = trajectory_similarity(manifold18, manifold22)

print("Trajectory similarity:", similarity)
```

```python

# Calculate Velocity (Step-to-step distance)
def get_velocity(m):
    return np.sqrt(np.sum(np.diff(m, axis=0)**2, axis=1))

v18 = get_velocity(manifold18)
v22 = get_velocity(manifold22)

# Slice to match lengths if necessary, and remove the startup spike
start_tr = 10
v1 = v18[start_tr:]
v2 = v22[start_tr:]

# Synchrony Calculation
sync_score = np.corrcoef(v1, v2)[0, 1]

plt.figure(figsize=(14, 5))
plt.plot(v1, label='Sub-18 Speed', color='red', alpha=0.6)
plt.plot(v2, label='Sub-22 Speed', color='blue', alpha=0.6)
plt.title(f"Dynamical Synchrony (Intersubject Correlation: {sync_score:.3f})")
plt.xlabel("Time (adjusted TRs)")
plt.ylabel("Velocity")
plt.legend()
plt.show()

print(f"The dynamical synchrony score is: {sync_score:.4f}")
```

```python
import pandas as pd

# 1. Smooth the manifolds (window size of 5 TRs = 10 seconds)
smooth_m18 = pd.DataFrame(manifold18).rolling(window=5, center=True).mean().values
smooth_m22 = pd.DataFrame(manifold22).rolling(window=5, center=True).mean().values

# 2. Re-calculate velocity on smoothed data
v18_smooth = np.sqrt(np.sum(np.diff(smooth_m18, axis=0)**2, axis=1))
v22_smooth = np.sqrt(np.sum(np.diff(smooth_m22, axis=0)**2, axis=1))

# 3. Clean NaNs from rolling window and re-check correlation
mask = ~np.isnan(v18_smooth) & ~np.isnan(v22_smooth)
new_sync = np.corrcoef(v18_smooth[mask][10:], v22_smooth[mask][10:])[0, 1]

print(f"Smoothed Synchrony Score: {new_sync:.4f}")
```
